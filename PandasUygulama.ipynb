### **2.Dataframe**

Pandas DataFrame, sütun ve satırlardan oluşan iki boyutlu veri tablolarıdır.

```python
# liste isimli listeyi Dataframe e çevirelim.
liste=[1,2,3,"a","b","c"]
df=pd.DataFrame(liste)
df
```

```python
#Oluşan Dataframe e Veriler başlığını verelim.
df=pd.DataFrame(liste,columns=["Veriler"])
df
```

```python
# iki boyutlu diziyi DataFrame e çevirelim ve Başlıklarını Ad ve Yaş olarak belirleyelim.
data=[["Ayhan",40],["Emine",6],["Meral",32]]
df=pd.DataFrame(data,columns=(["Ad","Yaş"]))
df
```

### **4.Concat işlemleri**

```python
rastgele1=np.random.randint(1,20,size=(4,3))
rastgele2=np.random.randint(20,40,size=(4,3))
```

```python
df1=pd.DataFrame(rastgele1,columns=["x","y","z"])
df1
```

```python
df2=pd.DataFrame(rastgele2,columns=["x","y","z"])
df2
```

```python
df=pd.concat([df1,df2]) # concat birleştirme işlemleri yapar.
df
```

```python
df=pd.concat([df1,df2], ignore_index=True)  #ignore_index ile birleştirilen listelerin index değerleri düzenlenir.
df
```

```python
df2.columns=["x","y","a"]
df=pd.concat([df1,df2],ignore_index=True)
df
```

```python
df=pd.concat([df1,df2],ignore_index=True, join="inner") # ortak alanlar yazılır.
df
```

### **6. Pandas Gruplama ve Kümeleme**

```python
data=pd.read_csv('/content/drive/MyDrive/2025/3-NumpyPandas/Pandas/Iris.csv')
data
```

```python
data.info()
```

```python
# Türlere göre gruplamak için groupby() yöntemini kullanarak verileri gruplandırınız.
data.groupby("Species")
```

```python
# Gruplanan türlerin sayısını öğrenmek için count fonksiyonunu kullanınız.
data.groupby("Species").count()
```

```python
# Gruplanan türlerin ortalamasını öğrenmek için mean fonksiyonunu kullanınız.
data.groupby("Species").mean()
```

```python
#Gruplanan türlerin toplamını öğrenmek için sum fonksiyonunu kullanınız.
data.groupby("Species").sum()
```

```python
# Gruplanan türlerin birden fazla özelliğine göre gruplamak için aggregate fonksiyonunu kullanınız.
data.groupby("Species").agg([min,max,np.std])
```

```python
# Gruplanan türlerin sadece belirli sütununa göre seçim yapmak için sütun kısmını seçme işlemini yapınız.
data.groupby("Species")["SepalLengthCm"].aggregate([max,min])
```
