### 1.Veri Setini Tanıma

* Matplotlib ile görselleştime temellerini öğrenmeye başlamadan önce, görselleştirme yapacağımız veri setini tanımakla başlayalım.
* Iris çiçek veri setini kullanacağız.
* Iris çiçek veri seti veya Fisher Iris veri seti, 1936 tarihli makalesinde İngiliz istatistikçi, öjeni uzmanı ve biyolog Ronald Fisher tarafından tanıtılan çok değişkenli bir veri setidir.
* İçerisinde 3 farklı sınıf bulunmaktadır. Bunlar setosa, versicolor ve virginica'dır.

```python
import pandas as pd
veri=pd.read_csv("/content/drive/MyDrive/2025/AdanZyeYapayZekaEgitimleri2025/Educations/3-NumpyPandas/Matplotlib/iris.csv")
veri.head()
```

* Her bir satır, bir çiçeğe ait ölçüm değerlerini gösterir. Özellikler sırasıyla sepal-length (alt yaprak uzunluğu cm), sepal-with (alt yaprak genişliği cm), pedal-length (üst yaprak genişliği), pedal-width (üst yaprak uzunluğu). Sınıflarımız ise setosa, versicolor ve virginica.

```python
veri.describe()
```

### 2.Çizgi Grafiği

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(veri["Id"],veri["SepalLengthCm"],color="red",alpha=0.7)
plt.title("Alt Yaprak Uzunluğu")
plt.xlabel("id")
plt.ylabel("cm")
plt.grid(True)
plt.show()
#
```

### 3.Saçılım Grafiği

* Saçılım diyagramı, iki farklı değişkenin arasındaki ilişkiyi belirlemek için kullanılır. Aralarındaki ilişkinin sebebi görülemese de, ilgili iki değişkenin arasında direkt olarak bir ilişki bulunup bulunmadığı ve bu ilişkinin ne derece güçlü olduğu görülebilir.

```python
plt.figure()
plt.scatter(veri["SepalLengthCm"],veri["SepalWidthCm"],color="blue",s=15,alpha=0.7) # alpha = saydamlık, s = size
plt.title("Alt Yaprak Uzunluğu ve Alt Yaprak Genişliği")
plt.xlabel("Alt Yaprak Uzunluğu")
plt.ylabel("Alt Yaprak Genişliği")
plt.grid(True)
```

### 3.Histogram Grafiği

* Histogram, gruplandırılmış bir veri dağılımının sütun grafiğiyle gösterimidir. Diğer bir ifadeyle, tekrarlı sayılardan oluşan verilerin, uygulanan işlemlerden sonra önce tabloya, tablodan yararlanarak grafiğe aktarılması, yani veri gruplarının grafiğinin dikdörtgen sütunlar halinde gösterilmesidir.

```python
# şimdi sepal length ve petal length sütunlarının histogramı çizdirelim
plt.figure() # figür oluştur - dış çerçeve
plt.hist(veri["SepalLengthCm"], color = "green", alpha = 0.5, bins = 20, label = "Alt Yaprak Uzunluk") # alpha = saydamlık, bins = bar sayısı
plt.hist(veri["PetalLengthCm"], color = "orange", alpha = 0.5, bins = 20, label = "Üst Yaprak Uzunluk") # alpha = saydamlık, bins = bar sayısı
plt.title('Alt Yaprak Uzunluğu ve Üst Yaprak Uzunluğu - Histogram')
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend() # label'ların görünebilmesi için
plt.show()
```

### 4.Çubuk Grafiği

* Çubuk grafik kategorik verileri temsil ettikleri değerlerle orantılı yükseklik veya uzunluktaki dikdörtgen çubuklarla sunan bir grafiktir.
* Çubuklar dikey veya yatay olarak çizilebilir.

```python
sutunlar = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
tanım = veri.describe()
ortalama = tanım.iloc[1,[1,2,3,4]]
plt.figure() # figür oluştur - dış çerçeve
plt.bar(sutunlar,ortalama )
plt.title('Alt-Üst Yaprak Uzunluğu ve Genişliği Ortalama')
plt.ylabel("Cm")
plt.show()
```

### 5.Alt Plan Grafiği

* Alt plan grafiklerinde birden fazla grafik tek bir figür/çerçeve üzerinde bulunur.

```python
fig, ax = plt.subplots(2, 1) # figür üzerine 2*1 lik bir tablo oluştur. bu tablo üzerine ax (eksen) ekle
fig.suptitle("Alt Yaprak Uzunluk ve Üst Yaprak Uzunluk", fontsize=14)
ax[0].plot(veri.Id,veri.SepalLengthCm,color="red",label= "SepalLengthCm")
ax[0].set_title('Alt Yaprak Uzunluk') # başlık
ax[0].set_xlabel("id") # x ekseni label
ax[0].set_ylabel("Cm") # y ekseni label

ax[1].plot(veri.Id,veri.PetalLengthCm,color="green",label= "PetalLengthCm")
ax[1].set_title('Üst Yaprak Uzunluk')
ax[1].set_xlabel("id")
ax[1].set_ylabel("Cm")

fig.tight_layout() # figür arası yerleşimi ayarla
fig.subplots_adjust(top=0.85)
plt.show()
```
